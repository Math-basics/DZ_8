#### Задание 1

Монетку подбрасывают 15 раз, найдите вероятность выпадения простого числа орлов.

В данном эксперименте максимально возможное количество орлов равно количеству бросков, т. е. 15. Простые числа меньше 15: 2, 3, 5, 7, 11, 13

В каждом броске вероятность выпадения орла или решки = 0,5. Количество всех возможных исходов 2 ** 15 = 32768. Количество событий, при которых выпадает ровно n орлов находим из комбинаторики:$$\frac{15!}{n!(15- n)!}$$ 
Для получения итогового ответа суммируем вероятности выпадения орлов простых чисел 2, 3, 5, 7, 11, 13.

```python
# Простые числа
n_simple = [2, 3, 5, 7, 11, 13]
# Количество всех возможных исходов
N = 2 ** 15
# Итоговый результат
result = 0

for n in n_simple:
    P = (np.math.factorial(15) / (np.math.factorial(n) * np.math.factorial(15 - n))) / N
    print(f'Вероятность выпадения {n} орлов составляет {P}')
    result += P         
print(f'\nВероятность выпадения простого числа орлов составляет {result}')
```

```python
# Проверим себя!

# Простые числа
n_simple = [2, 3, 5, 7, 11, 13]

# Количество попыток
trials = 100000
# Количество выпадений простых чисел орлов
result = 0

for i in range(trials):
    # Подбрасываем монетку 15 раз
    current_set = list(np.random.randint(2, size = 15)) 
    # Пусть орёл соответствует нулю
    if current_set.count(0) in n_simple:
        result += 1
print(f'Из {trials} попыток выпало {result} простых чисел орлов, что соответствует вероятности {result / trials}')
print('Результаты эксперимента близки к расчётным!')
```

```python
# Нормальное распределение

# Создаём поле для 5 графиков
fig, axes = plt.subplots(nrows = 1, # Одна строка с графиками
                         ncols = 5, # В строке 5 графиков
                         figsize=(18, 5), # Размер общего поля
                         sharey = 'all' # Делаем вертикальную ось общей для всех графиков
                        ) 

axes[0].hist(np.random.normal(loc = 0, scale = 1, size = 100000),
             bins = np.arange(-5, 5, 0.1), # Создаём список диапазонов от 0 до 10 с шагом 1
             density=True
             )
axes[0].set_xlim(-5, 5)
axes[0].set_title('loc = 0, scale = 1', fontsize = 'x-large')

axes[1].hist(np.random.normal(loc = 1, scale = 1, size = 100000),
             bins = np.arange(-5, 5, 0.1), # Создаём список диапазонов от 0 до 10 с шагом 1
             density=True
             )
axes[1].set_xlim(-5, 5)
axes[1].set_title('loc = 1, scale = 1', fontsize = 'x-large')

axes[2].hist(np.random.normal(loc = -1, scale = 1, size = 100000),
             bins = np.arange(-5, 5, 0.1), # Создаём список диапазонов от 0 до 10 с шагом 1
             density=True
             )
axes[2].set_xlim(-5, 5)
axes[2].set_title('loc = -1, scale = 1', fontsize = 'x-large')

axes[3].hist(np.random.normal(loc = 0, scale = 2, size = 100000),
             bins = np.arange(-5, 5, 0.1), # Создаём список диапазонов от 0 до 10 с шагом 1
             density=True
             )
axes[3].set_xlim(-5, 5)
axes[3].set_title('loc = 0, scale = 2', fontsize = 'x-large')

axes[4].hist(np.random.normal(loc = 0, scale = 0.5, size = 100000),
             bins = np.arange(-5, 5, 0.1), # Создаём список диапазонов от 0 до 10 с шагом 1
             density=True
             )
axes[4].set_xlim(-5, 5)
axes[4].set_title('loc = 0, scale = 0.5', fontsize = 'x-large')

# Общая надпись    
fig.suptitle('Гистограммы плотности распределения вероятности для нормального распределения', fontsize = 'xx-large')  
# Надпись по вертикальной оси
axes[0].set_ylabel('Плотности распределения\nвероятности', fontsize = 'xx-large')
# Надпись по горизонтальной оси (присваиваем среднему графику для расположения по центру)
axes[2].set_xlabel('Диапазоны значений', fontsize = 'xx-large')  

plt.show()
```
